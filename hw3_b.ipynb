```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
S = 100+1
V = np.zeros(S)
opt_A = np.zeros(S)
error = 1e-4
Delta = 1
delta = 1
p_h = 0.25

def p(s, a):
    prob = np.zeros(S)
    prob[s+a] = p_h
    prob[s-a] = 1-p_h
    return prob

def r(s, a):
    return 1 if s+a==100 or s==100 else 0

while Delta > error:
    Delta = 0
    for s in range(1, S):
        v = V[s]
        # actions:
        A = np.arange(min(s, S-s))
        r_V = np.zeros(len(A))
        for i,a in enumerate(A): 
            r_V[i] =  (r(s,a)+delta*V).dot(p(s,a))

        V[s] = np.max(r_V)
        opt_A[s] = np.argmax(r_V)
        Delta = max(Delta, abs(V[s] - v))

fig, ax = plt.subplots(dpi=300)
ax.plot(np.arange(S-1), V[:-1])
ax.set_title('value function')
plt.show()
print(opt_A)
        
```
